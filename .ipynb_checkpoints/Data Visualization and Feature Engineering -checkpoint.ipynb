```python
iris = pd.read_csv("iris.csv", names=["Id", "SepalLength", "SepalWidth", "PetalLength", "PetalWidth", "Species"])

print(iris.head())

import matplotlib.pyplot as plt # import library with plotting stuff
import matplotlib.colors as mcolors # import colors stuff to use hex color values

# create color directory
colors = {"Iris-setosa": mcolors.to_hex("#e612b4"), "Iris-versicolor":mcolors.to_hex("#58a8fc"), "Iris-virginica":mcolors.to_hex("#eeff38")}

figure, axis = plt.subplots()

# scatter each plot (petallength against petalwidth)
for i in range(len(iris["PetalLength"])): 
    axis.scatter(iris["PetalLength"][i], iris["PetalWidth"][i],color=colors[iris["Species"][i]])
    
# set title and exis names
axis.set_title("Petal length vs. petal width")
axis.set_xlabel("Petal_Length")
axis.set_ylabel("Petal_Width")
```

```python
# Now starts the feature engineering part of the lab
import pandas as panda 
import matplotlib.pyplot as plt 

dataframe = panda.read_csv("iris.csv",names=["Id", "SepalLength", "SepalWidth", "PetalLength", "PetalWidth", "Species"])

dataframe.head()
```

```python
dataframe = panda.read_csv("winemag-data-130k-v2.csv") # import wine csv
dataframe.head() # show top 5

# create histogram on 'points' column
plt.hist(dataframe.points, bins=10, rwidth=0.8) 
plt.xlabel('Points') 
plt.ylabel('Count') 
plt.show()
```

```python
thresold_max = dataframe['price'].quantile(0.90) # max value

thresold_min = dataframe['price'].quantile(0.10) # min value

# visualize the data below the minimum thresholds
dataframe[dataframe['price'] < thresold_min]


```

```python
# create a new dataframe which includes the data (POINTS, not price lol, looks better) between the thresholds we just made
new_dataframe = dataframe[(dataframe['points'] < thresold_max) & (dataframe['points'] > thresold_min)]

# show and compare
new_dataframe.shape
#dataframe.shape
```

```python
# create histogram on 'price' column from dataframe
plt.hist(dataframe.points, bins=10, rwidth=0.8) 
plt.title("Old dataframe")
plt.xlabel('Points') 
plt.ylabel('Count') 
plt.show()

# create histogram on 'price' column from new_dataframe
plt.hist(new_dataframe.points, bins=10, rwidth=0.8)
plt.title("New dataframe")
plt.xlabel('Points') 
plt.ylabel('Count') 
plt.show()

## We see that there are less outliers (at each end of the table)
```
